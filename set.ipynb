A set in Python is a collection of unique and unordered data items. It is defined using curly braces {} and the elements are separated by commas.

Here's an example of how to create a set in Python:


my_set = {1, 2, 3}
Use code with caution
This creates a set named my_set with three elements: 1, 2, and 3.

Here are some important characteristics of sets in Python:

Uniqueness: Sets only store unique elements. If you try to add duplicate elements, they will be ignored.
Unordered: The elements in a set are not ordered. This means that you cannot access elements by their index.
Membership: You can check if an element is present in a set using the in operator.
Here are some common operations you can perform on sets in Python:

Adding elements: You can add elements to a set using the add() method or the update() method.
Removing elements: You can remove elements from a set using the remove() method or the discard() method.
Union: You can combine two sets into a new set using the union() method or the | operator.
Intersection: You can find the common elements of two sets using the intersection() method or the & operator.
Difference: You can find the elements that are in one set but not in another set using the difference() method or the - operator.
