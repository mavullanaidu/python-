```python
#append
my_list=['list',12,23,34.56]
my_list.append(144)
print(my_list)
# extend
my_list=['list',123,12.3, -34]
print(my_list)
my_list.extend(['pythom',124,56])
print(my_list)




```

```python
# insert
my_list=[12,-343, 56, 'mylist']
print(my_list)
my_list.insert(3, 'gaa')
print(my_list)
```

```python
# extend
my_list=[12,23.4,'mylist','python']
print(my_list)
my_list.remove(12)
print(my_list)
```

```python
#pop
my_list=[12,34.4,'python','data']
print(my_list)
my_list.pop()
print(my_list)
```

```python
# revome
my_list=[12,-12,-12.2,12.4,'revome','data']
print(my_list)
my_list.remove('data')
print(my_list)
```

```python
# count
my_list=[12,34,12,'mavulla naidu','guna','guna','guna']
b=my_list.count('guna')
print(my_list)
print(b)
```

```python
# index
my_list=['mavulla', 'naidu']
b=my_list.index('naidu')
print(my_list)
print(b)
```

```python
# reverse
my_list=[12,34,56.67,678.8,'ghgg','ghh']
my_list.reverse()
print(my_list)
```

```python
# sort
my_list=[12,13,45,67,89,9,-12,4.5,7]
my_list.sort()
print(my_list)
```

```python
my_list=[12,13,45,67,89,9,-12,4.5,7]
my_list.sort(reverse=True)
print(my_list)
```

```python
# copy
my_list=[12,13,45,67,89,9,-12,4.5,7]
b=my_list
print(b)
print(my_list)
my_list.append(567)
print(my_list)
```

```python
#clear
my_list=[12,13,45,67,89,9,-12,4.5,7]
my_list.clear()
print(my_list)
```
