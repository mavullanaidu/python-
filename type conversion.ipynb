```python
# indexing
# it is calling from one charecter in the string
# they are two types
# forward indexing ()it is start with 0
# backward indexing() it is start with -1

my_str='python data structure'
print(my_str)
my_str1='1234567890'
print(my_str[1])
print(my_str[2])
print(my_str1[3])
print(my_str1[-1])
print(my_str[-3])
```

```python
# slicing
#syntax(start:end)
# start:
#:end
#:
# negative slicing
my_string = 'okkasari vacchi povachukadhe'
my_string1 = '14357'
print(my_string[:])
print(my_string1[:])
print(my_string[2:])
print(my_string1[:2])
print(my_string1[1:5])
print(my_string[-1:])
print(my_string[:-1])
print(my_string[-1:-9])
```

```python
# skipping
#syntax [start:end:step]
my_string='o chiina ramulamma'
print(my_string[::2])
my_string2='12345678'
print(my_string2[::2])
print(my_string2[::-1])
print(my_string[::-2])
```
